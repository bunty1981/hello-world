```python
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords

str = "This is a sample string with multiple words not a valid string - so go on."
en_stopwords = set(stopwords.words('english'))
#print("These are the stopwords: ", en_stopwords)

str_no_stopwords = ' '.join ([word for word in str.split(' ') if word.lower() not in en_stopwords])
print("This is the original string:", str)
print("This is the string without stopwords:", str_no_stopwords)

# We will now modify the stopwords
en_stopwords.remove('not')
en_stopwords.add('go')

str_no_stopwords = ' '.join ([word for word in str.split(' ') if word.lower() not in en_stopwords])
print("This is the original string:", str)
print("This is the string without the *modified* stopwords:", str_no_stopwords)
```

```python
# This section focuses on the regular expression (re) functions in Python
import re
str = "This is a sample string with multiple words not a valid string - so go on."
path_str  = "C:\\Users\\username\\Documents\\file.txt" # double backslash is used to escape the backslash character
print("Path str: ", path_str)

path_str = r"C:\Users\username\Documents\file.txt" # raw string is used to avoid escaping the backslash character
print("Path str: ", path_str)

# Search function in re module - finds the first occurrence of the pattern in the string
match = re.search('\D', path_str) #\D matches all non-digit characters
match = re.search(r'\\D', path_str) #r == raw data, \\ == backslash character (double backlash used to escape the backslash character), \D == all non-digit characters
if match:
    print("The string is: ", path_str)
    print("Match object: ", match)
    print("Search function found: ", match.group())
else:
    print("Search function found: ", None)  # None is returned if the pattern is not found



```

```python
# More uses of re module
from posixpath import sep
import re

reviews = ['sam was a great help to me in the store', 
 'the cashier was very rude to me, I think her name was eleanor', 
 'amazing work from sadeen!', 
 'sarah was able to help me find the items i needed quickly', 
 'lucy is such a great addition to the team', 
 'great service from sara she found me what i wanted']

# Find all reviews for 'sara' or 'sarah'
sara_reviews = [review for review in reviews if re.search(r'sara|sarah', review)] # the | operator is like OR in regex

print("Reviews for 'sara' or 'sarah' using the '|' operator:", *sara_reviews, sep='\n')
print('\n')

sara_reviews = [review for review in reviews if re.search(r'sarah?', review)] # The ? is used to make 'h' optional in Sarah
print("Reviews for 'sara' or 'sarah' using the '?' operator:", *sara_reviews, sep='\n')
```

```python
# A few more regexp examples
import re

reviews = ['sam was a great help to me in the store', 
 'the cashier was very rude to me, I think her name was eleanor', 
 'amazing work from sadeen!', 
 'sarah was able to help me find the items i needed quickly', 
 'lucy is such a great addition to the team', 
 'great service from sara she found me what i wanted']

# All reviews starting with an 'a'
a_reviews = [review for review in reviews if re.search(r'^a', review)]  # ^ used to match the start of a line
print("Reviews starting with 'a':", *a_reviews, sep='\n')

# All reviews ending with 'y'
y_reviews = [review for review in reviews if re.search(r'y$', review)]  # $ used to match the end of a line
print("Reviews ending with 'y':", *y_reviews, sep='\n')
print('\n')

# Reviews without punctuation
punctuation = r'[^\w\s]'
no_punct_reviews = [re.sub(punctuation, " ", review) for review in reviews]  # \w matches any word character (alphanumeric and underscore), \s matches any whitespace character

# The below will NOT work, as the string replace function can only handle literal strings, not regex patterns 
# no_punct_reviews = [review.replace(punctuation, " ") for review in reviews]

print("Reviews without punctuation:", *no_punct_reviews, sep='\n')
```
