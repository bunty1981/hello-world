```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import pandas as pd
import re

""" # Download necessary NLTK data
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet') """

# Load the dataset
df = pd.read_csv('tripadvisor_hotel_reviews.csv')
df.info()
df.head()

##df = df[0:10]  # Select a subset of the data for processing

# Preprocess the text data, assuming each element in the 'Review' column is a string
def preprocess_text(text):
    ## Remove HTML tags
    #text = re.sub(r'<.*?>', '', text)
    
    # Convert to lower case
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Convert asterisk (*) to the word star in the context of hotel ratings, so that it's not removed by punctuation removal
    tokens = [re.sub(r'[*]', 'star', word) for word in tokens]

    # Remove stop words & strip punctuation
    stop_words = set(stopwords.words('english'))
    stop_words.remove('not')  # Keep 'not' as it's important for sentiment analysis
    tokens = [re.sub(r'[^\w\s]', '', word) for word in tokens if word not in stop_words]
    # # print(tokens)

    # Remove empty tokens
    tokens = [word for word in tokens if word != '']
    # # print(tokens)

    # Lemmatize the text
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Join the tokens back into a single string
    text = ' '.join(tokens)
    return text

# Apply the preprocessing function to the 'Review' column
df['preprocessed_text'] = df['Review'].apply(preprocess_text)

# Display the 1st 5 rows of the preprocessed data
df.head(5)

##--------------------------------------------------------------

# # Alternate method to apply preprocessing with a overall function call
# df['lower_case'] = df['Review'].str.lower()

# # Remove stop words
# stop_words = set(stopwords.words('english'))
# stop_words.remove('not')  # Keep 'not' as it's important for sentiment analysis
# df['stop_words_removed'] = df['lower_case'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in stop_words]))

# # Convert asterisk (*) to the word star
# #df['stop_words_removed'] = df['stop_words_removed'].str.replace('\*', 'star') # example of using the string.replace function
# df['stop_words_removed'] = df.apply(lambda x: re.sub(r'[*]', 'star', x['stop_words_removed']), axis = 1) # example of traversing and modifying the dataframe row-wise

# # Remove punctuation
# df['cleaned_text'] = df['stop_words_removed'].apply(lambda x: re.sub(r'[^\w\s]', '', x)) 

# # Tokenize the text
# df['tokenized_text'] = df['cleaned_text'].apply(lambda x: word_tokenize(x))
                                            
# # Lemmatize the text & join back into a single string
# lemmatizer = WordNetLemmatizer()
# df['lemmatized_text'] = df['tokenized_text'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x]))

# # Comparing the outputs of the two methods (preprocess_text function vs. alternate method above)
# df['compare_preprocessing'] = df['preprocessed_text'] == df['lemmatized_text']

# print( "Number of matches: ", df['compare_preprocessing'].sum() )
# print("Number of elements: ", df['compare_preprocessing'].count())




```
